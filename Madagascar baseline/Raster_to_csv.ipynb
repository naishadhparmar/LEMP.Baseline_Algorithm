```
merged_data = Raster('merged_raster.tif')
```

```
df = merged_data.to_pandas(max_pixels=25000000)
df.shape
columns = ['X', 'Y', 'Esta', 'Gain', 'Loss', 'Roads', 'Waterways', 'Elevation', 'Slope']
df.columns = columns
df.head()

df.loc[df.Loss == 100.0, 'Loss'] = 0 #100 means persistent non-forest which is essentially no loss
df.loc[df.Loss != 0 , 'Loss'] = 1 #All non-zero losses are pixels which faced loss at one point in time

df.to_csv('merged_raster_as_csv.csv', index=False)
```
